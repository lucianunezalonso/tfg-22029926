```python
adoptados = df.loc[df['Nombre de ultimo estado'] == 'Adoptado']
```

```python
adoptados
```

```python
def clasificar_especie(tipo):
    if tipo == 'Europeo Común':
        return 'gato'
    else:
        return 'perro'
```

```python
adoptados['Especie'] = adoptados['Tipo de animal'].apply(clasificar_especie)
```
