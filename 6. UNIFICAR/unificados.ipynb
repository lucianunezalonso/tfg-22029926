```python
df = pd.concat([andalucia, asturias, aragon, baleares, canarias, castillalamancha,
               castillaleon, cantabria, cataluña, valencia, madrid, euskadi, larioja,
               murcia, navarra], axis=0)
```

```python
df
```

```python
df.drop(['C.P.','LOCALIDAD','CONCEJO','PROVINCIA'], axis=1, inplace=True)
```

```python
canina= ['canino','canina','perr']
felina= ['felin','gat']
```

```python
def categorizar(nombre):
    if any(palabra in str(nombre).lower() for palabra in canina):
        return 'canina'
    elif any(palabra in str(nombre).lower() for palabra in felina):
        return 'felina'
    else:
        return 'mixta'
```

```python
df['Categoria'] = df['Nombre'].apply(categorizar)
```
