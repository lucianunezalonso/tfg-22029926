```python
from selenium import webdriver
import time

from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import urllib.request
from PIL import Image

import pandas as pd

path= "C:\Program Files (x86)\chrimedriver.exe"
```

```python
# Abrir una instancia del navegador
driver = webdriver.Chrome(path)
# Va a la p√°gina que le digo
driver.get("https://www.protectoramalaga.com/perros-en-adopcion/1/")
```

```python
# FUNCIONA OK

# Boton de cookies
cookies_button = driver.find_element("xpath","/html/body/div[1]/div/a")
cookies_button.click()

data = []

while True:
    driver.refresh()

    # Obtiene los elementos de la p√°gina actual (es una lista por p√°gina)
    elementos = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, "//*[@class= 'w-box inverse']")))
    
    i = 0
    while i < len(elementos):
        
        elemento = WebDriverWait(driver, 10).until(EC.visibility_of_all_elements_located((By.XPATH, "//*[@class= 'w-box inverse']")))[i]
        
        nombre_xpath = elemento.find_element("xpath", ".//h2/a")
        nombre = nombre_xpath.text
        
        imagen_elemento = elemento.find_element("xpath",".//a/img")
        imagen_url = imagen_elemento.get_attribute("src")

        
        print(nombre)
        
        
        nombre_xpath.click()
        
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//section[@class='slice bg-3']")))
        
        # Extraer informaci√≥n de la p√°gina a la que se dirige
        sexo = driver.find_element("xpath", ".//div/div/div/div[1]/div/div[2]/div[1]/h4").text
        raza= driver.find_element("xpath", ".//div/div/div/div[1]/div/div[2]/div[2]/h4").text
        edad= driver.find_element("xpath", ".//div/div/div/div[1]/div/div[2]/div[3]/h4").text
        tama√±o= driver.find_element("xpath", ".//div/div/div/div[1]/div/div[2]/div[4]/h4").text
        try:
            descripcion= driver.find_element("xpath", "//div/div/div/div[1]/div/div[2]/div[7]").text
        except NoSuchElementException:
            descripcion= driver.find_element("xpath", "//div/div/div/div[1]/div/div[2]/div[6]").text
            
        # Agregar la informaci√≥n a la lista
        data.append([nombre, sexo, raza, edad, tama√±o, descripcion, imagen_url])
    
        driver.back()
        
        i += 1
        
        time.sleep(2)

    try:
        # Busca el bot√≥n de "siguiente"
        siguiente = driver.find_element("xpath", "//div[@class='pull-right']/ul/li[@class='next'][2]/a")
        driver.execute_script("arguments[0].click();", siguiente)
        print("---------SIGUIENTE P√ÅGINA---------")
    except NoSuchElementException:
        print("---------SE ACAB√ì---------")
        # El ciclo termina
        break
```

```python
# Crear un dataframe a partir de la lista de datos
df = pd.DataFrame(data, columns=['Nombre', 'Sexo', 'Raza', 'Edad', 'Tama√±o', 'Descripcion', 'Imagen'])
```
