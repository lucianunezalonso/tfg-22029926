```python
# Me quedo solo con la de dirección y la de ccaa
df = df.loc[:, ['Nombre', 'Direccion', 'ccaa','latitud','longitud']]
```

1. TELÉFONO

```python
from faker import Faker
fake = Faker('es_ES')
```

```python
df['Telefono'] = [fake.phone_number() for _ in range(len(df))]
```

2. CORREO

```python
import re
import random
import string
```

```python
df['Nombre_limpio'] = df['Nombre'].apply(lambda x: re.sub('[^a-zA-Z0-9@.+_]+', '', x.lower()))
```

```python
def generate_email(name):
    domain = fake.domain_name()
    username = re.sub(r'\W+', '', name) # eliminar caracteres especiales del nombre
    length = random.randint(5, 10) # generar una longitud aleatoria
    return f"{username[:length]}@{domain}"
```

```python
df['Correo'] = df['Nombre_limpio'].apply(generate_email)
```

```python
df
```

3. URL

```python
df['url'] = df['Nombre_limpio'].apply(lambda x: f"http://www.{x.replace(' ', '').lower()}.com")
```

```python
df
```

4. FACEBOOK

```python
df['Facebook'] = df['Nombre_limpio'].apply(lambda x: fake.uri_page() + "/" + x.lower().replace(" ", "-"))
```
