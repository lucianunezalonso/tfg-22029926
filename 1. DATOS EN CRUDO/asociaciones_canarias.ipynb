```python
import pandas as pd
import ezodf
```

```python
doc = ezodf.opendoc('asociaciones_canarias.ods')
```

```python
hoja = doc.sheets[0]
```

```python
datos = []
for fila in hoja.rows():
    fila_datos = []
    for celda in fila:
        fila_datos.append(celda.value)
    datos.append(fila_datos)
```

```python
df = pd.DataFrame(datos[1:], columns=datos[0])
```

```python
# Guardar DataFrame en un archivo de Excel
df.to_excel('asociaciones_canarias.xlsx', index=False)
```
