PARA OPTIMIZAR MUCHO TIEMPO:

Voy a crear una función, que dada una palabra, me extraiga del dataframe todos los registros que la contengan en su denominación

```python
def extraer_por_palabra(df, palabra):
    df_palabra = df[df['DENOMINACION'].str.lower().str.contains(palabra.lower())]
    return df_palabra
```

```python
animal= extraer_por_palabra(df_otros, 'animal')
animal
```

PASO EL OUTPUT DE DF_OTROS A DF, PORQUE ME QUIERO QUEDAR CON ÉL:

Creo una función que me saque de df_otros, la información que ya esté en el df generado

```python
def restar_df(df_otros, df_palabra):
        df_otros = df_otros[~df_otros.isin(df_palabra)].dropna()
        return df_otros
```
