```python
def eliminar_filas_listanegra(df, listanegra):
    # Crea una Serie booleana que indica si cada fila contiene alguna palabra de la lista negra
    mascara = df['Nombre Asociación'].str.contains('|'.join(listanegra), case=False).fillna(False)
    # Elimina las filas que cumplen la condición anterior
    df = df[~mascara]
    return df
```

```python
listanegra=['salm','canar','pajar','caball','equin','exot','voluntar',
           'animalista','cuidador','tor','gall','captura','ornit',
           'avi','eq','maltrat','sostenib','proye','estudi','observ',
           'horse','burro','ecues','cavall','palom','bird','adiestr',
           'cinof','educ','dep','salvamento','agility']
```

```python
list(df['Nombre Asociación'])
```

```python
df= eliminar_filas_listanegra(df, listanegra)
```
