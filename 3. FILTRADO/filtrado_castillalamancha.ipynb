FUNCIONES:

```python
# Crea un df con los registros en los que aparece el nombre que le digo
def extraer_por_palabra(df, palabra):
    df_palabra = df[df['DENOMINACION'].str.lower().str.contains(palabra.lower())]
    return df_palabra
```

```python
def restar_df(df_minuendo, df_sustraendo):
        df_diferencia = df_minuendo[~df_minuendo.isin(df_sustraendo)].dropna()
        return df_diferencia
```

```python
def eliminar_filas_listanegra(df, listanegra):
    # Crea una Serie booleana que indica si cada fila contiene alguna palabra de la lista negra
    mascara = df['DENOMINACION'].str.contains('|'.join(listanegra), case=False)
    # Elimina las filas que cumplen la condición anterior
    df = df[~mascara]
    return df
```
