Lo que voy a ir haciendo es aplicar máscaras al dataframe de manera que finalmente me quede con un nuevo conjunto de datos solamente con los registros que me interesan: CENTROS DE PROTECCIÓN ANIMAL

FUNCIONES:

```python
def extraer_por_palabra(df, palabra):
    df_palabra = df[df['ASOCIACIÓN'].str.lower().str.contains(palabra.lower())]
    return df_palabra
```

```python
def restar_df(df_minuendo, df_sustraendo):
        df_diferencia = df_minuendo[~df_minuendo.isin(df_sustraendo)].dropna()
        return df_diferencia
```

```python
def eliminar_filas_listanegra(df, listanegra):
    # Crea una Serie booleana que indica si cada fila contiene alguna palabra de la lista negra
    mascara = df['ASOCIACIÓN'].str.contains('|'.join(listanegra), case=False)
    # Elimina las filas que cumplen la condición anterior
    df = df[~mascara]
    return df
```
