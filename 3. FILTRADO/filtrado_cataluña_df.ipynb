```python
def eliminar_filas_listanegra(df, listanegra):
    # Crea una Serie booleana que indica si cada fila contiene alguna palabra de la lista negra
    mascara = df['nom_entitat'].str.contains('|'.join(listanegra), case=False)
    # Elimina las filas que cumplen la condición anterior
    df = df[~mascara]
    return df
```

```python
def extraer_por_palabra(df, palabra):
    df_palabra = df[df['nom_entitat'].str.lower().str.contains(palabra.lower())]
    return df_palabra
```

```python
def restar_df(df_minuendo, df_sustraendo):
    df_diferencia = df_minuendo.merge(df_sustraendo, how='outer', indicator=True)
    df_diferencia = df_diferencia.loc[df_diferencia['_merge'] == 'left_only'].drop(columns='_merge')
    return df_diferencia
```
