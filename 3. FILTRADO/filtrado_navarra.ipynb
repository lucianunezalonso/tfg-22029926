```python
# Crea un df con los registros en los que aparece el nombre que le digo
def extraer_por_palabra(df, palabra):
    df_palabra = df[df['DENOMINACIÓN'].str.lower().str.contains(palabra.lower())]
    return df_palabra
```

```python
def restar_df(df_minuendo, df_sustraendo):
        df_diferencia = df_minuendo[~df_minuendo.isin(df_sustraendo)].dropna()
        return df_diferencia
```

```python
def eliminar_filas_listanegra(df, listanegra):
    # Crea una Serie booleana que indica si cada fila contiene alguna palabra de la lista negra
    mascara = df['DENOMINACIÓN'].str.contains('|'.join(listanegra), case=False)
    # Elimina las filas que cumplen la condición anterior
    df = df[~mascara]
    return df
```

```python
print(len(list(df['DENOMINACIÓN'].unique())))
list(df['DENOMINACIÓN'].unique())
```

```python
listanegra= ['caza','pesca','acuar','natur','ornit','micologi','eco','eko','biolog',
            'cria','ganad','ambien','caball','buho','propiet','ciencia','taur','tor',
            'bird','anguil','salvaj','natur','COORD','PEÑA','socie','energ','bucef',
            'bonsa','agilit','verde','urberri','arbol','huerto','igualda','agua','huert',
            'obed','igual','agua','bison','laguna','transport','deport','rio','cañada',
            'tierr','zorr','clima','patrimo','explor','educ','mosca','sociocult','canto',
            'recic']
```
